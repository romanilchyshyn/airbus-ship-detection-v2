```python
import pandas as pd

df = pd.read_csv('../data/train_ship_segmentations_v2.csv')
df
```

```python
empty = df
```

```python
empty = df[df['EncodedPixels'].apply(lambda x: all(p == '' for p in x))]
empty
```

```python
non_empty = df[df['EncodedPixels'].apply(lambda x: any(p != '' for p in x))]
non_empty
```

```python
sample = 1000

n_empty = int(sample * 0.3)
n_non_empty = sample - n_empty

            # sample from each
empty_sample = empty.sample(n=min(n_empty, len(empty)))
non_empty_sample = non_empty.sample(n=min(n_non_empty, len(non_empty)))

empty_sample
```

```python
non_empty_sample
```

```python
df = pd.concat([empty_sample, non_empty_sample]).sample(frac=1).reset_index()
df
```
